```python
# Two columns from the table are not needed 
# currency_id
# mp_commoditysource
# Unnamed: 0
# They need to be dropped
```

```python
drop_columns = ['Unnamed: 0', 'currency_id', 'mp_commoditysource']
df = df.drop(columns = drop_columns)
```

```python
# Check for null values
```

```python
df.isnull().sum()
```

```python
# replace null values in the 'state' column with 'not specified'
```

```python
df["state"].fillna("not specified", inplace = True)
```

```python
df.isnull().sum() # this means the dataset no longer has any empty column
```

```python
df['produce']
```

```python
# split column 'produce' 

# split the produce column in two
new = df["produce"].str.split ("-", n =1, expand = True)
# rename first column to products
df["products"] = new[0]
# rename the second column to retail
df["retail"] = new[1]
# drop produce and retail column
df.drop(columns = ["produce"], inplace = True)
# drop retail column
df.drop(columns = ["retail"], inplace = True)
```
